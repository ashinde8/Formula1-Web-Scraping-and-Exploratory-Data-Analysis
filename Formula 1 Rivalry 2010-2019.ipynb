```python
def rivalry(year1,year2):
     
    Hamilton_wins = []
    Vettel_wins = []
    Other_drivers = []
    
    Mercedes_wins = []
    Ferrari_wins = []
    Red_Bull_wins = []
    Others_Constructors = []
    
    dataframe = []
    
    for i in range(year1,year2):

        yr_col = []
        RB = ['RBR Renault','Red Bull Racing Renault', 'Red Bull Racing TAG Heuer','Red Bull Racing Honda']
        
        yrs = str(i)
        file_name = yrs + '_Race_Results.csv'
        df = pd.read_csv(file_name)
        
        for i in range(len(df['CONSTRUCTOR'])):
            
            if df['CONSTRUCTOR'][i] in RB:
                df['CONSTRUCTOR'][i] = df['CONSTRUCTOR'][i].replace(df['CONSTRUCTOR'][i],'Red Bull Racing')
        
        dataframe.append(df)

    # The list dataframe consists of all the datasets from the year 2010 to 2019
    
    for df in dataframe:
        
        Ham = 0
        Vet = 0
        Alo = 0
        Oth_drvr = 0
        
        Merc = 0
        Ferr = 0
        RedBull = 0
        Oth_Cnstr = 0
        
        for i in df['CONSTRUCTOR']:
            
            if i == 'Mercedes':
                Merc += 1
            elif i == 'Ferrari':
                Ferr += 1
            elif i == 'Red Bull Racing':
                RedBull += 1
            else:
                Oth_Cnstr += 1
        
        Mercedes_wins.append(Merc)
        Ferrari_wins.append(Ferr)
        Red_Bull_wins.append(RedBull)
        Others_Constructors.append(Oth_Cnstr)
        
        
        for j in df['WINNER']:
            
            if j == 'Hamilton':
                Ham += 1
            elif j == 'Vettel':
                Vet += 1
            else:
                Oth_drvr += 1
                
        Hamilton_wins.append(Ham)
        Vettel_wins.append(Vet)
        Other_drivers.append(Oth_drvr)
        
    return Mercedes_wins, Ferrari_wins, Red_Bull_wins, Others_Constructors, Hamilton_wins, Vettel_wins, Other_drivers
                
```

```python
result = rivalry(2010,2020)   
```
